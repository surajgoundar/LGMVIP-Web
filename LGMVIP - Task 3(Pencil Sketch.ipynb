```python
# Importing the Libraries

import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Load the Image

image=cv2.imread("Arijit Singh.png")
```

```python
# Read the image in RGB format

im_rgb=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10,6))
plt.imshow(im_rgb)
plt.title("Original Image")
plt.axis("off");
```

```python
# Convert the original image to grayscale image

gray_image=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(10,6))
plt.imshow(gray_image,cmap="gray")
plt.title("GrayScale Image")
plt.axis("off");
```

```python
# Inverting image 

inverted_image=255-gray_image
plt.figure(figsize=(10,6))
plt.imshow(inverted_image,cmap="gray")
plt.title("Inverted Image")
plt.axis("off");
```

```python
# Blur image

blur=cv2.GaussianBlur(inverted_image,(21,21),0)
plt.figure(figsize=(10,6))
plt.imshow(blur,cmap="gray")
plt.title("Smoothen Image")
plt.axis("off");
```

```python
# Convert image to Pencil Sketch

sketch=cv2.divide(gray_image,255-blur,scale=255)
plt.figure(figsize=(10,6))
plt.imshow(sketch,cmap="gray")
plt.title("Final Sketch Image")
plt.axis("off");
```
